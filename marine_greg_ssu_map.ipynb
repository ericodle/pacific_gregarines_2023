```python
import pandas as pd

import folium

url = 'https://raw.githubusercontent.com/ericodle/pacific_gregarines_2023/main/marine_greg_ncbi_spreadsheet.csv'

df = pd.read_csv(url)

# Remove rows with non-convertible values or replace them with valid coordinates
df = df[pd.to_numeric(df['Latitude'], errors='coerce').notnull()]
df = df[pd.to_numeric(df['Longitude'], errors='coerce').notnull()]

# Create a folium map centered around the first coordinate
map = folium.Map(location=[df['Latitude'].iloc[0], df['Longitude'].iloc[0]], zoom_start=10)

# Add markers to the map for each coordinate pair
for index, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']],
                  tooltip=row['Species']).add_to(map)

# Display the map
map
```
